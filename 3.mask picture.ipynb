```python
import os, re
import numpy as np
from PIL import Image

from epub2txt import epub2txt
from ultralytics import YOLO, SAM
import torch
from transformers import CLIPProcessor, CLIPModel

# =========================
# 已帮你填好路径
# =========================
EPUB_PATH = r"D:\巴企蠕动中\skill 总\skill python\3.book drawing line\inputs\epubs\Science and the Modern World — Alfred North Whitehead.epub"

IMAGES_DIR = r"D:\巴企蠕动中\skill 总\skill python\3.book drawing line\inputs\image"

OUT_DIR = r"D:\巴企蠕动中\skill 总\skill python\3.book drawing line\outputs 3"
# =========================

os.makedirs(OUT_DIR, exist_ok=True)

# ---------- 1) 读 EPUB 做词频 ----------
print("Reading EPUB…")
text = epub2txt(EPUB_PATH).lower()

word_freq = {}
for w in re.findall(r"[a-zA-Z]+", text):
    word_freq[w] = word_freq.get(w, 0) + 1
print("✅ EPUB loaded.")

# ---------- 2) 加载模型 ----------
print("Loading models…")
yolo = YOLO("yolov8x-seg.pt")   # 第一次会自动下载
sam  = SAM("sam_b.pt")         # 第一次会自动下载

clip = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
clip_proc = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
print("✅ Models loaded.")

# ---------- CLIP 标签候选 ----------
TEXT_PROMPTS = [
    "painting", "wall", "sky", "tree", "floor", "ceiling", "bench",
    "person", "window", "glass", "building", "panel", "interior",
    "art", "sculpture", "light", "shadow", "vegetation", "plant",
    "architecture", "object", "furniture", "pillar", "column",
    "door", "frame", "chair", "display", "screen", "landscape", "room"
]

def clip_label(pil_img, box_xyxy):
    x1, y1, x2, y2 = map(int, box_xyxy)
    crop = pil_img.crop((x1, y1, x2, y2))
    inputs = clip_proc(text=TEXT_PROMPTS, images=crop, return_tensors="pt", padding=True)
    with torch.no_grad():
        logits = clip(**inputs).logits_per_image[0]
    return TEXT_PROMPTS[int(logits.argmax().item())].lower()

# ---------- 3) 处理图片 ----------
print("\nProcessing images…")
imgs = [f for f in os.listdir(IMAGES_DIR) if f.lower().endswith((".jpg", ".jpeg", ".png"))]

if not imgs:
    raise RuntimeError(f"❌ 图片文件夹里没有 jpg/jpeg/png：{IMAGES_DIR}")

for name in imgs:
    path = os.path.join(IMAGES_DIR, name)
    print("→", name)

    pil = Image.open(path).convert("RGB")
    img = np.array(pil)

    # YOLO 检测
    y = yolo(img)[0]
    boxes = y.boxes.xyxy.cpu().numpy() if y.boxes is not None else []

    # SAM 分割
    s = sam.predict(source=img, retina_masks=True)[0]
    masks = s.masks.data.cpu().numpy() if s.masks is not None else []

    keep = np.zeros(img.shape[:2], dtype=np.uint8)

    for box in boxes:
        x1, y1, x2, y2 = map(int, box)
        if x2 <= x1 or y2 <= y1:
            continue

        label = clip_label(pil, box)
        if word_freq.get(label, 0) > 5:
            for m in masks:
                if np.any(m[y1:y2, x1:x2] > 0.5):
                    keep[m > 0.5] = 1

    out = img.copy()
    out[keep == 0] = [255, 255, 255]

    save_path = os.path.join(OUT_DIR, f"masked_{name}")
    Image.fromarray(out).save(save_path)
    print("Saved:", save_path)

print("\n✅ Done.")
print("Output folder:", OUT_DIR)
```

```python
# --- 把阈值先降到 1，避免“书里没提到”导致全白 ---
FREQ_THRESHOLD = 1

print("\nProcessing images…")
imgs = [f for f in os.listdir(IMAGES_DIR) if f.lower().endswith((".jpg", ".jpeg", ".png"))]

for name in imgs:
    path = os.path.join(IMAGES_DIR, name)
    print("\n==============================")
    print("→", name)

    pil = Image.open(path).convert("RGB")
    img = np.array(pil)

    y = yolo.predict(img, conf=0.15, iou=0.45, imgsz=960, verbose=False)[0]
    boxes = y.boxes.xyxy.cpu().numpy() if y.boxes is not None else []

    if len(boxes) == 0:
        print("⚠️ YOLO: 0 detections -> 这张图一定会白（没框就没保留对象）")
        continue

    print(f"YOLO boxes: {len(boxes)}")

    s = sam.predict(source=img, retina_masks=True, verbose=False)[0]
    masks = s.masks.data.cpu().numpy() if s.masks is not None else []

    keep = np.zeros(img.shape[:2], dtype=np.uint8)
    kept_any = False

    for i, box in enumerate(boxes):
        label = clip_label(pil, box)
        freq = word_freq.get(label, 0)
        print(f"[{i}] CLIP label='{label}' | freq_in_book={freq}")

        if freq > FREQ_THRESHOLD:
            # 找与框有重叠的 mask
            for m in masks:
                x1, y1, x2, y2 = map(int, box)
                if np.any(m[y1:y2, x1:x2] > 0.5):
                    keep[m > 0.5] = 1
                    kept_any = True

    if not kept_any:
        print("⚠️ 最终没有任何区域被保留：要么书里词频不够，要么框与 mask 没重叠。输出会是纯白。")

    out = img.copy()
    out[keep == 0] = [255, 255, 255]

    save_path = os.path.join(OUT_DIR, f"masked_{name}")
    Image.fromarray(out).save(save_path)
    print("Saved:", save_path)
```
