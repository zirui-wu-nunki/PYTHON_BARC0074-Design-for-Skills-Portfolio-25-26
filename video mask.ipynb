```python
from moviepy.editor import VideoFileClip, ColorClip

def safe_write_video(input_path, output_path, start_time=None, end_time=None, fps=30):
    # 1️⃣ 打开视频
    video = VideoFileClip(input_path)

    # 2️⃣ 裁剪
    if start_time is not None or end_time is not None:
        subclip = video.subclip(start_time or 0, end_time or video.duration)
    else:
        subclip = video

    # 3️⃣ mask 转 RGB
    if subclip.ismask:
        subclip = subclip.to_RGB()

    # 4️⃣ 确保 fps、size、duration 都有
    if not hasattr(subclip, "fps") or subclip.fps is None:
        subclip.fps = fps
    if not hasattr(subclip, "size") or subclip.size is None:
        subclip.size = video.size
    if not hasattr(subclip, "duration") or subclip.duration is None:
        subclip.duration = subclip.end - subclip.start

    # 5️⃣ 处理音频
    if subclip.audio is None:
        subclip = subclip.set_audio(None)

    # 6️⃣ 写出视频
    print(f"🎬 正在生成视频: {output_path}")
    subclip.write_videofile(
        output_path,
        codec="libx264",
        audio_codec="aac",
        fps=subclip.fps,
        threads=4
    )

    video.close()
    print("✅ 视频生成完成")

# 使用示例
safe_write_video(
    r"D:\巴企蠕动中\skill python\video-classification-3d-cnn-pytorch\Input_films\hong ri.mp4",
    r"D:\巴企蠕动中\skill python\video-classification-3d-cnn-pytorch\output_films\scene1_hong_ri.mp4",
    start_time=0,
    end_time=10,
    fps=30
)
```

```python
def find_best_match(reference_features, target_features):
    best_start = 0
    best_score = float('inf')

    for i in range(len(target_features) - len(reference_features)):
        distance = 0
        for j in range(len(reference_features)):
            d = np.linalg.norm(
                target_features[i + j] - reference_features[j]
            )
            distance += d

        if distance < best_score:
            best_score = distance
            best_start = i

    return best_start, best_score
```

```python
import os, json, glob
import numpy as np
from moviepy.editor import VideoFileClip
from PIL import Image

PROJECT = r"D:\巴企蠕动中\skill python\video-classification-3d-cnn-pytorch"
os.chdir(PROJECT)

# 1) 读取最新 features json
features_path = sorted(glob.glob("*features*.json"), key=os.path.getmtime, reverse=True)[0]
with open(features_path, "r", encoding="utf-8", errors="replace") as f:
    data = json.load(f)

videos = data["videos"]
vA = videos[0]
vB = videos[1]

A_path = vA["video"]
B_path = vB["video"]
A_feats = vA["features"]
B_feats = vB["features"]

print("[INFO] A:", A_path, "windows =", len(A_feats))
print("[INFO] B:", B_path, "windows =", len(B_feats))

# 2) 取出向量矩阵
A = np.array([x["vec"] for x in A_feats], dtype=np.float32)
B = np.array([x["vec"] for x in B_feats], dtype=np.float32)

# 3) 余弦相似度准备
def normalize(x):
    n = np.linalg.norm(x, axis=1, keepdims=True) + 1e-8
    return x / n

A_n = normalize(A)
B_n = normalize(B)

# 4) 滑窗：在 B 中找一段长度=len(A) 的窗口，平均相似度最高
L = len(A_n)
if L == 0 or len(B_n) < L:
    raise ValueError("windows 数量不够（A=0 或 B<A）。先在 Step1 确认 windows 不为 0。")

best_i = 0
best_score = -1e9

# 用点积求每个位置的平均相似
for i in range(0, len(B_n) - L + 1):
    score = float((A_n * B_n[i:i+L]).sum(axis=1).mean())
    if score > best_score:
        best_score = score
        best_i = i

best_start = float(B_feats[best_i]["t0"])
best_end = float(B_feats[best_i + L - 1]["t1"])

print(f"[OK] best match in B: i={best_i}, score={best_score:.4f}")
print(f"[OK] time range in B: {best_start:.2f}s -> {best_end:.2f}s")

# 5) subclip 切片输出
out_dir = os.path.join(PROJECT, "matched_output")
os.makedirs(out_dir, exist_ok=True)

matched_video_path = os.path.join(out_dir, "SecondMatchedFirstVideo.mp4")
with VideoFileClip(B_path) as clip:
    sub = clip.subclip(best_start, best_end)
    sub.write_videofile(matched_video_path, codec="libx264", audio_codec="aac")

print("[OK] saved matched clip ->", matched_video_path)

# 6) 抽帧（每 0.5 秒）
frames_dir = os.path.join(out_dir, "frames")
os.makedirs(frames_dir, exist_ok=True)

with VideoFileClip(matched_video_path) as mclip:
    t = 0.0
    step = 0.5
    idx = 1
    while t < mclip.duration:
        frame = mclip.get_frame(t)  # RGB ndarray
        img = Image.fromarray(frame)
        img.save(os.path.join(frames_dir, f"frame_{idx:04d}.png"))
        idx += 1
        t += step

print("[OK] frames saved ->", frames_dir, "count =", idx-1)
```

```python
import os, json, glob
import numpy as np
from moviepy.editor import VideoFileClip
from PIL import Image

PROJECT = r"D:\巴企蠕动中\skill python\video-classification-3d-cnn-pytorch"
os.chdir(PROJECT)

# 1) 读取最新 features
features_path = sorted(glob.glob("*features*.json"), key=os.path.getmtime, reverse=True)[0]
with open(features_path, "r", encoding="utf-8", errors="replace") as f:
    data = json.load(f)

videos = data["videos"]
vA = videos[0]
vB = videos[1]

A_path = vA["video"]
B_path = vB["video"]
A_feats = vA["features"]
B_feats = vB["features"]

print("[INFO] features file:", features_path)
print("[INFO] A:", os.path.basename(A_path), "windows =", len(A_feats))
print("[INFO] B:", os.path.basename(B_path), "windows =", len(B_feats))

if len(A_feats) == 0 or len(B_feats) == 0:
    raise ValueError("A 或 B 的 windows = 0。说明 feature 提取没成功产出窗口。下一步我会给你一段调整参数重跑 features 的代码。")

A = np.array([x["vec"] for x in A_feats], dtype=np.float32)
B = np.array([x["vec"] for x in B_feats], dtype=np.float32)

def normalize(x):
    n = np.linalg.norm(x, axis=1, keepdims=True) + 1e-8
    return x / n

A_n = normalize(A)
B_n = normalize(B)

# 2) 选择查询长度 Q：取两边都能容纳的长度，最多 12 个 windows
Q = min(12, len(A_n), len(B_n))
if Q < 2:
    raise ValueError(f"windows 太少，Q={Q}。至少需要 2 个 windows 才能匹配。")

# 3) 在 A 里也滑动（找“哪一段 A”在 B 里最像）
best = {
    "score": -1e9,
    "a_i": 0,
    "b_i": 0
}

# 预先把 B 的每个滑窗段拿出来会占内存，这里用直接循环（数据不大）
for a_i in range(0, len(A_n) - Q + 1):
    A_seg = A_n[a_i:a_i+Q]
    for b_i in range(0, len(B_n) - Q + 1):
        score = float((A_seg * B_n[b_i:b_i+Q]).sum(axis=1).mean())
        if score > best["score"]:
            best = {"score": score, "a_i": a_i, "b_i": b_i}

a_i = best["a_i"]
b_i = best["b_i"]
best_score = best["score"]

A_start = float(A_feats[a_i]["t0"])
A_end   = float(A_feats[a_i + Q - 1]["t1"])
B_start = float(B_feats[b_i]["t0"])
B_end   = float(B_feats[b_i + Q - 1]["t1"])

print(f"[OK] best score = {best_score:.4f}")
print(f"[OK] A segment: i={a_i}  {A_start:.2f}s -> {A_end:.2f}s")
print(f"[OK] B segment: i={b_i}  {B_start:.2f}s -> {B_end:.2f}s")

# 4) subclip 切 B 段
out_dir = os.path.join(PROJECT, "matched_output")
os.makedirs(out_dir, exist_ok=True)

matched_video_path = os.path.join(out_dir, "SecondMatchedFirstVideo.mp4")
with VideoFileClip(B_path) as clip:
    sub = clip.subclip(B_start, B_end)
    sub.write_videofile(matched_video_path, codec="libx264", audio_codec="aac")

print("[OK] saved matched clip ->", matched_video_path)

# 5) 抽帧（每 0.5 秒）
frames_dir = os.path.join(out_dir, "frames")
os.makedirs(frames_dir, exist_ok=True)

with VideoFileClip(matched_video_path) as mclip:
    t = 0.0
    step = 0.5
    idx = 1
    while t < mclip.duration:
        frame = mclip.get_frame(t)
        Image.fromarray(frame).save(os.path.join(frames_dir, f"frame_{idx:04d}.png"))
        idx += 1
        t += step

print("[OK] frames saved ->", frames_dir, "count =", idx-1)
```

```python
import os, glob
import cv2
import numpy as np

PROJECT = r"D:\巴企蠕动中\skill python\video-classification-3d-cnn-pytorch"
frames_dir = os.path.join(PROJECT, "matched_output", "frames")

# 你要查询的图片：把下面改成你想查的那一张（例如 frame_0013.png）
QUERY_IMG = os.path.join(frames_dir, "frame_0013.png")

# 抽帧的间隔（你之前是每 0.5 秒一张）
FRAME_STEP_SEC = 0.5

def hist_feature(path):
    img = cv2.imread(path)
    if img is None:
        raise FileNotFoundError(path)
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hist = cv2.calcHist([hsv], [0, 1], None, [50, 60], [0, 180, 0, 256])
    cv2.normalize(hist, hist)
    return hist.flatten()

def similarity(h1, h2):
    # 越大越相似（1.0 最高）
    return float(cv2.compareHist(h1.astype(np.float32), h2.astype(np.float32), cv2.HISTCMP_CORREL))

# build index
all_frames = sorted(glob.glob(os.path.join(frames_dir, "*.png")))
q_feat = hist_feature(QUERY_IMG)

scores = []
for fp in all_frames:
    f = hist_feature(fp)
    s = similarity(q_feat, f)
    scores.append((s, fp))

scores.sort(reverse=True, key=lambda x: x[0])

print("[QUERY]", os.path.basename(QUERY_IMG))
print("\nTop 10 similar frames:")
for rank, (s, fp) in enumerate(scores[:10], start=1):
    name = os.path.basename(fp)
    # frame_0001 -> t=0.0, frame_0002 -> t=0.5 ...
    idx = int(os.path.splitext(name)[0].split("_")[1])
    t = (idx - 1) * FRAME_STEP_SEC
    print(f"{rank:02d}. score={s:.4f}  {name}  approx_time_in_matched_clip={t:.2f}s")
```

```python
import os, glob
import numpy as np
import cv2
from PIL import Image

PROJECT = r"D:\巴企蠕动中\skill python\video-classification-3d-cnn-pytorch"
frames_dir = os.path.join(PROJECT, "matched_output", "frames")
out_dir = os.path.join(PROJECT, "matched_output", "sam_color_overlay")
os.makedirs(out_dir, exist_ok=True)

# 1) 自动拿到真实存在的帧文件
frame_files = sorted(glob.glob(os.path.join(frames_dir, "*.png")))
if not frame_files:
    raise FileNotFoundError(f"[ERROR] frames_dir 里没有 png：{frames_dir}")

print("总帧数:", len(frame_files))
print("前 5 个:", [os.path.basename(x) for x in frame_files[:5]])
print("后 5 个:", [os.path.basename(x) for x in frame_files[-5:]])

# ✅ 选一张帧：默认第一张；你也可以改成 frame_files[12] 之类
frame_path = frame_files[0]
FRAME_NAME = os.path.basename(frame_path)
stem = os.path.splitext(FRAME_NAME)[0]
print("\n当前选择帧:", FRAME_NAME)

# 2) 读原图
base = np.array(Image.open(frame_path).convert("RGB")).astype(np.float32) / 255.0

# 3) 找对应 masks
mask_folder = os.path.join(PROJECT, "matched_output", "sam_masks", "masks", stem)
mask_paths = sorted(glob.glob(os.path.join(mask_folder, "mask_*.png")))

if not mask_paths:
    raise FileNotFoundError(
        f"[ERROR] 找不到 masks：{mask_folder}\n"
        f"请检查你的 masks 目录结构是否是 .../sam_masks/masks/{stem}/mask_*.png"
    )

print("mask 数量:", len(mask_paths))
print("mask 示例:", os.path.basename(mask_paths[0]))

def color_for(i):
    # HSV 里均匀跳色（稳定、区分明显）
    h = (i * 37) % 180
    hsv = np.uint8([[[h, 220, 255]]])
    rgb = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)[0, 0].astype(np.float32) / 255.0
    return rgb

# ============ A) Multiply / 正片叠底（更像你说的“正片叠底叠出来”） ============
strength = 0.70  # 0~1 越大颜色越“压”越明显
out_mul = base.copy()

for i, mp in enumerate(mask_paths):
    m = (np.array(Image.open(mp).convert("L")) > 0)
    col = color_for(i)  # 0..1
    blended = out_mul * col  # multiply
    out_mul[m] = (1 - strength) * out_mul[m] + strength * blended[m]

save_mul = os.path.join(out_dir, f"{stem}_multiply.png")
Image.fromarray(np.clip(out_mul * 255, 0, 255).astype(np.uint8)).save(save_mul)
print("[OK] saved multiply ->", save_mul)

# ============ B) Alpha 叠色（更像“彩色分区涂抹”） ============
alpha = 0.45  # 0~1 越大颜色越显眼
out_alpha = base.copy()

for i, mp in enumerate(mask_paths):
    m = (np.array(Image.open(mp).convert("L")) > 0)
    col = color_for(i)
    out_alpha[m] = (1 - alpha) * out_alpha[m] + alpha * col

save_alpha = os.path.join(out_dir, f"{stem}_alpha.png")
Image.fromarray(np.clip(out_alpha * 255, 0, 255).astype(np.uint8)).save(save_alpha)
print("[OK] saved alpha ->", save_alpha)
```

```python
import os, glob
import numpy as np
import cv2
from PIL import Image

# ✅ 你的项目根目录（按你刚给的新路径）
PROJECT = r"D:\巴企蠕动中\skill 总\skill python\video-classification-3d-cnn-pytorch"
MATCHED = os.path.join(PROJECT, "matched_output")

MASKS_ROOT = os.path.join(MATCHED, "sam_masks", "masks")  # 这里是 frame_0001...frame_0026 的父目录

# 输出目录
OUT_SEG = os.path.join(MATCHED, "sam_color_seg")          # 纯彩色分区（透明背景）
OUT_MUL = os.path.join(MATCHED, "sam_color_multiply")     # 正片叠底到原图
OUT_ALP = os.path.join(MATCHED, "sam_color_alpha")        # alpha 叠色到原图
os.makedirs(OUT_SEG, exist_ok=True)
os.makedirs(OUT_MUL, exist_ok=True)
os.makedirs(OUT_ALP, exist_ok=True)

def color_for(i: int):
    # 颜色稳定、区分度高
    h = (i * 37) % 180
    hsv = np.uint8([[[h, 220, 255]]])
    rgb = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)[0, 0].astype(np.float32) / 255.0
    return rgb  # 0..1

def find_base_frame(frame_stem: str):
    """
    在 matched_output 里尽量找到 frame_0001 对应的原图帧：
    先查常见目录，再递归查找
    """
    candidates = []

    # 常见位置（你也可以按自己的结构继续加）
    common_dirs = [
        os.path.join(MATCHED, "frames"),
        os.path.join(MATCHED, "matched_frames"),
        os.path.join(MATCHED, "input_frames"),
        os.path.join(MATCHED, "frames_rgb"),
        MATCHED,
    ]

    exts = [".png", ".jpg", ".jpeg", ".webp"]
    for d in common_dirs:
        for ext in exts:
            p = os.path.join(d, frame_stem + ext)
            if os.path.exists(p):
                return p

    # 递归搜（可能会慢一点，但你的量不大 26 帧，OK）
    patterns = []
    for ext in exts:
        patterns.append(os.path.join(MATCHED, "**", frame_stem + ext))
    for pat in patterns:
        candidates += glob.glob(pat, recursive=True)
    candidates = sorted(set(candidates))
    return candidates[0] if candidates else None

# 读取有哪些 frame 文件夹
frame_dirs = sorted([d for d in os.listdir(MASKS_ROOT) if os.path.isdir(os.path.join(MASKS_ROOT, d))])
print("发现 frame 文件夹数量:", len(frame_dirs))
print("示例:", frame_dirs[:5], "..." , frame_dirs[-5:])

# 叠加参数（你老师想要的重点是 Multiply）
multiply_strength = 0.70  # 正片叠底强度 0~1
alpha_strength = 0.45     # 普通彩色涂层强度 0~1

for frame_stem in frame_dirs:
    mask_folder = os.path.join(MASKS_ROOT, frame_stem)
    mask_paths = sorted(glob.glob(os.path.join(mask_folder, "mask_*.png")))
    if not mask_paths:
        print("[跳过] 没有 mask:", frame_stem)
        continue

    # 为了更像“一个个物件”叠在一起：可按面积从大到小排序（更稳定）
    masks = []
    for mp in mask_paths:
        m = (np.array(Image.open(mp).convert("L")) > 0)
        area = int(m.sum())
        masks.append((area, mp, m))
    masks.sort(key=lambda x: x[0], reverse=True)

    H, W = masks[0][2].shape

    # A) 先做“彩色分区透明图”（RGBA）
    seg_rgb = np.zeros((H, W, 3), dtype=np.float32)
    seg_a   = np.zeros((H, W), dtype=np.float32)

    for i, (area, mp, m) in enumerate(masks):
        col = color_for(i)
        # 让每个mask区域直接被颜色填充；alpha 为 1
        seg_rgb[m] = col
        seg_a[m] = 1.0

    seg_rgba = np.dstack([seg_rgb, seg_a])
    save_seg = os.path.join(OUT_SEG, f"{frame_stem}_seg_rgba.png")
    Image.fromarray(np.clip(seg_rgba * 255, 0, 255).astype(np.uint8), mode="RGBA").save(save_seg)

    # B) 尝试找原图帧：找到就做 Multiply / Alpha overlay
    base_path = find_base_frame(frame_stem)
    if base_path is None:
        print(f"[只有分区图] {frame_stem} 找不到原图帧（已输出 seg_rgba）")
        continue

    base = np.array(Image.open(base_path).convert("RGB")).astype(np.float32) / 255.0

    # Multiply 叠底
    out_mul = base.copy()
    for i, (area, mp, m) in enumerate(masks):
        col = color_for(i)
        blended = out_mul * col
        out_mul[m] = (1 - multiply_strength) * out_mul[m] + multiply_strength * blended[m]
    save_mul = os.path.join(OUT_MUL, f"{frame_stem}_multiply.png")
    Image.fromarray(np.clip(out_mul * 255, 0, 255).astype(np.uint8)).save(save_mul)

    # Alpha 叠色（更像彩色涂层）
    out_alp = base.copy()
    for i, (area, mp, m) in enumerate(masks):
        col = color_for(i)
        out_alp[m] = (1 - alpha_strength) * out_alp[m] + alpha_strength * col
    save_alp = os.path.join(OUT_ALP, f"{frame_stem}_alpha.png")
    Image.fromarray(np.clip(out_alp * 255, 0, 255).astype(np.uint8)).save(save_alp)

    print(f"[OK] {frame_stem} | seg_rgba + multiply + alpha")
    print("     原图:", base_path)

print("\n完成！输出目录：")
print("seg(透明分区):", OUT_SEG)
print("multiply(正片叠底):", OUT_MUL)
print("alpha(涂层叠色):", OUT_ALP)
```
