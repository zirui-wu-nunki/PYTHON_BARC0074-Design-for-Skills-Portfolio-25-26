```python
import time
import csv
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

options = Options()
driver = webdriver.Chrome(options=options)

# 打开 GetYourGuide 灵感页面
driver.get("https://www.getyourguide.com/explorer/travel-inspiration/?utm_source=chatgpt.com")
time.sleep(5)

data = []

# 固定只抓 3 个“分页”（如果没有分页，根据你页面结构调整）
for page in range(3):
    print(f"正在抓取第 {page + 1} 页")

    # 抓取每个文章块
    articles = driver.find_elements(By.CSS_SELECTOR, "article")
    for art in articles:
        try:
            # 标题
            title = art.find_element(By.CSS_SELECTOR, "h3").text.strip()
        except:
            title = "无标题"

        try:
            # 链接
            link = art.find_element(By.CSS_SELECTOR, "a").get_attribute("href")
        except:
            link = "无链接"

        try:
            # 封面图（如果有）
            image = art.find_element(By.CSS_SELECTOR, "img").get_attribute("src")
        except:
            image = "无图片"

        data.append([title, link, image])

    # 尝试翻页（如果分页按钮存在）
    try:
        next_btn = driver.find_element(By.CSS_SELECTOR, "a.pagination-next")
        next_btn.click()
        time.sleep(5)
    except:
        print("没有可用的下一页，停止抓取")
        break

# 保存 CSV
with open("getyourguide_travel_inspiration.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["title", "link", "image"])
    writer.writerows(data)

print("📌 抓取完成，已保存到 getyourguide_travel_inspiration.csv")
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 1️⃣ Chrome 设置
options = Options()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")

driver = webdriver.Chrome(options=options)

try:
    # 2️⃣ 搜索关键词（Nature + Artificial Intelligence）
    keyword = "Nature Artificial Intelligence"
    url = f"https://www.semanticscholar.org/search?q={keyword}"
    driver.get(url)

    # 3️⃣ 等待搜索结果加载
    wait = WebDriverWait(driver, 30)
    wait.until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "div.search-result"))
    )

    time.sleep(2)

    # 4️⃣ 获取论文列表
    papers = driver.find_elements(By.CSS_SELECTOR, "div.search-result")

    print(f"\n共抓取到 {len(papers)} 篇论文：\n")

    # 5️⃣ 逐条提取信息
    for idx, paper in enumerate(papers, start=1):
        try:
            # 标题
            title = paper.find_element(By.CSS_SELECTOR, "h2 a").text

            # 作者
            authors = paper.find_element(
                By.CSS_SELECTOR,
                "span.author-list"
            ).text

            # 年份（有些没有，防报错）
            try:
                year = paper.find_element(
                    By.CSS_SELECTOR,
                    "span[data-test-id='year']"
                ).text
            except:
                year = "N/A"

            print(f"{idx}.")
            print(f"标题: {title}")
            print(f"作者: {authors}")
            print(f"年份: {year}")
            print("-" * 50)

        except Exception as e:
            print(f"❌ 第 {idx} 条抓取失败：{e}")

finally:
    # 6️⃣ 不自动关浏览器，方便你观察
    print("\n爬取完成，可以手动关闭浏览器。")

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import time

# 1️⃣ Chrome 设置
options = Options()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")

driver = webdriver.Chrome(options=options)
wait = WebDriverWait(driver, 30)

# 2️⃣ Google Scholar 搜索
query = "Nature Artificial Intelligence"
url = f"https://scholar.google.com/scholar?q={query.replace(' ', '+')}"
driver.get(url)

# 3️⃣ 等搜索结果
wait.until(
    EC.presence_of_element_located((By.CSS_SELECTOR, "div.gs_r"))
)

time.sleep(2)

papers = []

results = driver.find_elements(By.CSS_SELECTOR, "div.gs_r")

for r in results[:20]:
    try:
        title = r.find_element(By.TAG_NAME, "h3").text
        info = r.find_element(By.CLASS_NAME, "gs_a").text
        papers.append([title, info])
    except:
        continue

# 4️⃣ 保存 CSV
with open("nature_artificial_intelligence_papers.csv", "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.writer(f)
    writer.writerow(["Title", "Authors / Year"])
    writer.writerows(papers)

print(f"✅ 成功抓取 {len(papers)} 篇论文")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import time

# 关键词列表
keywords = ["nature", "artificial", "artificial intelligence", "nature artificial"]

# Chrome 设置
options = Options()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")
driver = webdriver.Chrome(options=options)
wait = WebDriverWait(driver, 30)

papers = []

# 分页抓取
for page in range(0, 100, 10):  # Google Scholar 每页 10 条结果
    query = " OR ".join(keywords)
    url = f"https://scholar.google.com/scholar?q={query.replace(' ', '+')}&start={page}"
    driver.get(url)
    
    wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.gs_r")))
    time.sleep(2)  # 稍作等待
    
    results = driver.find_elements(By.CSS_SELECTOR, "div.gs_r")
    
    for r in results:
        try:
            title = r.find_element(By.TAG_NAME, "h3").text
            info = r.find_element(By.CLASS_NAME, "gs_a").text
            citation = 0
            try:
                citation_text = r.find_element(By.XPATH, ".//a[contains(text(),'Cited by')]").text
                citation = int(citation_text.split()[-1])
            except:
                pass
            
            # 匹配关键词
            matched = [k for k in keywords if k.lower() in title.lower()]
            
            papers.append([title, info, matched, citation])
        except:
            continue

    print(f"Page {page//10 + 1} done, total papers collected: {len(papers)}")
    time.sleep(3)  # 每页间隔 3 秒，避免封 IP

# 保存 CSV
with open("scholar_papers.csv", "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.writer(f)
    writer.writerow(["Title", "Authors / Year", "Matched Keywords", "Citations"])
    writer.writerows(papers)

print(f"✅ 完成抓取，总共 {len(papers)} 篇论文")
driver.quit()

```

```python
import requests
import csv
import time

# 📌 keywords you want to search for
keywords = ["nature", "artificial", "artificial intelligence"]

# OpenAlex search base endpoint
BASE_URL = "https://api.openalex.org/works"

papers = []

# OpenAlex allow pagination, let's get ~100 results
per_page = 25  #每页数量
pages = 4      #4页 → ~100 条

for page in range(pages):
    # ⭐ build search query (OR between keywords)
    kw_query = " OR ".join(keywords)
    params = {
        "filter": f"title.search:{kw_query}",
        "per-page": per_page,
        "page": page + 1
    }

    print(f"Fetching page {page+1}")
    response = requests.get(BASE_URL, params=params)

    if response.status_code != 200:
        print("⚠ Error", response.status_code)
        break

    data = response.json()

    for item in data.get("results", []):
        title = item.get("display_name", "")
        authors = ", ".join([a["author"]["display_name"] for a in item.get("authorships", [])])
        year = item.get("publication_year", "")
        citations = item.get("cited_by_count", 0)

        # check which keywords appear in title
        matched = [k for k in keywords if k.lower() in title.lower()]

        papers.append([title, authors, year, ";".join(matched), citations])

    time.sleep(1)  # avoid query throttle

print(f"📊 Fetched {len(papers)} papers.")

# save CSV
with open("openalex_papers.csv", "w", encoding="utf-8-sig", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Title", "Authors", "Year", "Matched Keywords", "Citations"])
    writer.writerows(papers)

print("✅ Done!")

```
